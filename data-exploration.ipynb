```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

## Numbers Representation

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_books"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_books", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.Figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "spearman correlaton"})
heatmap.set_title("Correlation Heatmap")
```

## Number of books that would need to be removed with missing values
#### which we found to be under 5 percent so it is not that much, so therefore we can delete it.

```python
books[(books["description"].isna()) |
      (books["num_pages"].isna()) |
      (books["average_rating"].isna()) | 
      (books["published_year"].isna())] 
#above is the code that shows the books with issing values
```

```python
missing_books = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) & 
      ~(books["published_year"].isna())] 


#above is the code for the new set of data that dont have the missing values.
```

```python
missing_books
```

```python
missing_books["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
missing_books
```

```python
missing_books["words_in_description"] = missing_books["description"].str.split().str.len()
```

```python
missing_books
```

```python
missing_books.loc[missing_books["words_in_description"].between(25, 34), "description"]
```

```python
missing_books_above_25_words = missing_books[missing_books["words_in_description"]>=25]
```

```python
missing_books_above_25_words
```

```python
missing_books_above_25_words["title_and_subtitle"] = (
    np.where(missing_books_above_25_words["subtitle"].isna(), missing_books_above_25_words["title"],
             missing_books_above_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
missing_books_above_25_words
```

```python
missing_books_above_25_words["tagged_description"] = missing_books_above_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
missing_books_above_25_words
```

```python
(
    missing_books_above_25_words
    .drop(["subtitle", "missing_description", "age_of_books", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```
